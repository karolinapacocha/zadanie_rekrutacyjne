```python
import pandas as pd
import numpy as np
import datetime

def zadanie_pierwsze(sciezka_pliku):
    data = pd.read_csv(sciezka_pliku)
    data['data_urodzenia'] = data['data_urodzenia'].apply(lambda x: datetime.datetime.strptime(x, '%d.%m.%Y'))
    print("Liczba osób urodzonych po 31 grudnia 1999r.: ", data.imie[data.data_urodzenia>'1999.12.31'].count())
    
    kobiety = [imie for imie in data.imie if imie.endswith('a')]
    output = []
    for kobieta in kobiety:
        if kobieta not in output:
            output.append(kobieta)
    print("Unikalne imiona kobiet: ", output)
```
